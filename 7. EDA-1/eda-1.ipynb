#### Categorical Features - Count Plot

```python
df['churn'].value_counts()
```

```python
df['customer service calls'].value_counts()
```

```python
# subplot(nrows, ncols, index, **kwargs)
plt.figure(1)

plt.subplots_adjust(hspace=0.4, wspace=0.7)

plt.subplot(121)
sns.countplot(x='churn', data=df)

plt.subplot(122)
sns.countplot(x='customer service calls', data=df)

plt.show()

# 211, 212

'''
We've used the hspace and wspace arguments of plt.subplots_adjust,\
which specify the spacing along the height and width of the figure
'''
```

### Summary

While the histograms, discussed above, and bar plots may look similar, there are several differences between them:
1. Histograms are best suited for looking at the distribution of numerical variables while bar plots are used for categorical features.
2. The values on the X-axis in the histogram are numerical; a bar plot can have any type of values on the X-axis: numbers, strings, booleans.

The left chart above vividly illustrates the imbalance in our target variable.

The bar plot for Customer service calls on the right gives a hint that the majority of customers resolve their problems in maximum 2-3 calls.

### Handling Missing Values

```python
df = pd.DataFrame({
    'FirstName': ['abc', np.nan, 'def', 'ghi', 'jkl'],
    'LastName': ['mno', np.nan, 'pqr', 'stu', 'vwx'],
    'Age': [12, np.nan, 34, 54, 24],
    'Sex': ['m', np.nan, 'f', 'f', 'm'],
    'preTestScore': [4.0, np.nan, np.nan, 2.0, 3.0],
    'postTestScore': [25.0, np.nan, np.nan, 62.0, 70.0]
})

df
```

```python
df_no_miss = df.dropna()
df_no_miss
```

```python
df_clean = df.dropna(how='all')
df_clean
```

```python
df['location'] = np.nan
df
```

```python
df_col_clean = df.dropna(how='all', axis=1)
df_col_clean
```

```python
df_all_clean = df_col_clean.dropna(how='all', axis=0)
df_all_clean
```

```python
df
```

```python
# Filling missing values with Zeros.

df.fillna(0)
```

```python
# Fill in missing in preTestScore with the mean

df['preTestScore'].fillna(df['preTestScore'].mean(), inplace=True)
df
```

```python
# Now lets do everything in order.

df = pd.DataFrame({
    'FirstName': ['abc', np.nan, 'def', 'ghi', 'jkl'],
    'LastName': ['mno', np.nan, 'pqr', 'stu', 'vwx'],
    'Age': [12, np.nan, 34, 54, 24],
    'Sex': ['m', np.nan, 'f', 'f', 'm'],
    'preTestScore': [4.0, np.nan, np.nan, 2.0, 3.0],
    'postTestScore': [25.0, np.nan, np.nan, 62.0, 70.0]
})

df
```

```python
df['location'] = np.nan
df
```

```python
df_col_clean = df.dropna(how='all', axis=1)
df_all_clean = df_col_clean.dropna(how='all', axis=0)

df_all_clean
```

```python
df_all_clean['preTestScore'].fillna(df_all_clean['preTestScore'].mean(), inplace=True)
df_all_clean['postTestScore'].fillna(df_all_clean['postTestScore'].mean(), inplace=True)

df_all_clean
```

#### Summary

We saw how to handle missing values. 
1. If complete row or a column is having missing values, than its better to remove it completely.
2. We saw how to fill missing values by Zeros and Mean(Imputation)
