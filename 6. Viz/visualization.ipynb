# Visualization (Numerical & Categorical Data)

## 1 Univariate Analysis (for Numerical Variables)
i.e Analyzing one variable at a time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lets try to create random normal distribution with 100 data points.

x = np.random.normal(size=100)

print(x.shape)
```

```python
sns.distplot(x)

# by default-> sns.distplot(x, hist=True, kde=True)
```

### 1.1 Histogram

<b>REFERENCE: </b>https://statistics.laerd.com/statistical-guides/understanding-histograms.php

```python
sns.distplot(x, kde=False, rug=True)

# This is basically a frequency plot
# Try 'rug=False'
```

### 1.2 Kernel Density Estimation (1D Dimensionality plot)

<b>REFERENCE: </b>http://www.mvstat.net/tduong/research/seminars/seminar-2001-05/

```python
sns.distplot(x, hist=False, rug=True)
```

```python
sns.kdeplot(x, shade=True)
```

```python
sns.kdeplot(x, label='original')
sns.kdeplot(x, bw=0.2, label='bw=0.2')
sns.kdeplot(x, bw=2, label='bw=2')
plt.legend()
```

## 2 Bivariate Analysis (for Numerical Variables)

```python
mean, cov = [0, 1], [(1, 0.5), (0.5, 1)]

data = np.random.multivariate_normal(mean, cov, 1000)

df = pd.DataFrame(data, columns=['x', 'y'])

print(df.shape)
```

```python
df.describe()
```

### 2.1 Scatter Plot
<b>REFERENCE: </b>http://mste.illinois.edu/courses/ci330ms/youtsey/scatterinfo.html

```python
df.plot(kind='scatter', x = 'x', y = 'y')
```

```python
sns.jointplot(x='x', y='y', data=df)
```

### 2.2 Hexbin Plots
<b>REFERENCE: </b>https://www.meccanismocomplesso.org/en/hexagonal-binning/

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style('white'):
    sns.jointplot(x = x, y = y, kind = 'hex', color = 'k')
```

### 2.3 Kernel Density Estimation (2D Density Plot)

```python
sns.jointplot(x='x', y='y', data=df, kind='kde')
```

```python
mean, cov = [0, 1], [(1, 1), (1, 1)]

data_1 = np.random.multivariate_normal(mean, cov, 1000)

df_1 = pd.DataFrame(data_1, columns=['x', 'y'])

print(df_1.shape)
```

```python
sns.jointplot(x='x', y='y', data=df_1, kind='kde')
```

### 2.4 Pair Plot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

### 3.2 Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

### 3.3 Boxplot

Percentile basically says where a point lies in a sorted list.<br />

Median-> 50th percentile<br />
Quantiles-> 25th, 50th, 75th, 100th percentile<br /><br />

75th percentile-> 75 percent of all the observations are less than 75th percentile.

Eg: CAT Rank

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
print(tips[tips.time=='Lunch'])
```

```python
tips.day[tips.time=='Lunch'].value_counts()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time', dodge=True)

# dodge -> When hue nesting is used, whether elements should be shifted along the categorical axis.
```

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.boxplot(x='day', y='total_bill', data=tips, \
            hue='weekend', dodge=False)
```

```python
tips['weekend'].describe()
```
